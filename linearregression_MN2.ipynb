#  Linear Regression


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
! kaggle datasets download brendan45774/SALARY
!unzip SALARY.zip
```

```python
df = pd.read_csv('/content/Salary.csv')
df
```

```python
# Features and target variable
X = df[['YearsExperience']]
y = df['Salary']

print(type(X))
print(type(y))
y
```

```python
# **Data Splitting**

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# **Model Building**

model = LinearRegression()
```

```python
# **Model Training**

# Train the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

print(y_pred)
print(y_test)
```

```python
# Plotting the regression line and data points
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X), color='red')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Linear Regression - YearsExperience vs Salary')
plt.show()
```

```python
# Calculate mean squared error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
no_years=int(input("Enter no. of years: "))
y_pred = model.predict([[no_years]])
y_pred
```

```python
df = pd.read_csv('/content/multiple_linear_regression_dataset.csv')
df
```

```python
# Define features and target
X = df[['age', 'experience']]
y = df['income']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Create and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)
```
